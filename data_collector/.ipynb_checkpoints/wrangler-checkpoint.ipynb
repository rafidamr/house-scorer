```python
df = pd.read_csv('../csv/house_price_raw_3.csv')
df.columns = ['PriceTitle', 'Location', 'Size', 'Agent', 'Link', 'Desc']

def clean_price(s):
    result = re.search(r'Rp.(.*?)/m²', s).group(1)
    return result

df
```

```python
def to_numeric_term(s: str):
    s = s.lower()
    if s == 'miliar': return 1e9
    if s == 'juta': return 1e6
    if s == 'ribu': return 1e3

def to_price_per_m2(df_inp):
    str_m2 = 'Rp (.*?) /m²'
    str_total = 'Rp (.*?) Total'
    df = df_inp['PriceTitle'].str.extract(f'{str_m2}|{str_total}')
    df = df.assign(price_m2=[0.0 for _ in range(len(df))])
    
    pm2 = df.loc[:,0]
    pm2_f : pd.DataFrame = pm2[pm2.notnull()]
    pm2_f = pm2_f.str.extract('(.*?) ([a-zA-Z]*)')
    pm2_f.loc[:,0] = pm2_f.loc[:,0].str.replace(',','.').astype(float)
    pm2_f.loc[:,1] = pm2_f.loc[:,1].apply(to_numeric_term)
    pm2_f = pm2_f.loc[:,0] * pm2_f.loc[:,1]
    df.loc[pm2.notnull(), 'price_m2'] = pm2_f

    size = df_inp[['Size','Link']]
    # size = size.str.extract('(.*?) /m²').astype(int)
    return size[size['Size'].isnull()].count()
    # ptotal = df.loc[:,1]
    # ptotal_f : pd.DataFrame = ptotal[ptotal.notnull()]
    # ptotal_f = ptotal_f.str.extract('(.*?) ([a-zA-Z]*)')
    # ptotal_f.loc[:,0] = ptotal_f.loc[:,0].str.replace(',','.').astype(float)
    # ptotal_f.loc[:,1] = ptotal_f.loc[:,1].apply(to_numeric_term)
    # ptotal_f = ptotal_f.loc[:,0] * ptotal_f.loc[:,1]
    # df.loc[ptotal.notnull(), 'price_m2'] = ptotal_f
    # return df

x = to_price_per_m2(df)
x
```
